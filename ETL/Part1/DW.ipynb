Removing any whitespaces

```python
#Removing whitespaces from the column names
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

#Removing Extra Whitespace from Whole DataFrame
def whitespace_remover(df):
    for i in df.columns:
        if df[i].dtype == 'object':
            df[i] = df[i].map(str.strip)
        else:
            pass
    return df

df1= whitespace_remover(df1)
df2= whitespace_remover(df2)

#df1_columnlist = df1.columns.tolist()
#df2_columnlist = df2.columns.tolist()
#ndf2
```

Adding the College Name(s) in the datasources

df1

```python
df1_majorlist = df1['Major'].tolist()
df1_collegenamelist= []

for i in range(len(df1_majorlist)):
  if df1_majorlist[i] == 'Computer Sc' or df1_majorlist[i] == 'Information Sc' or df1_majorlist[i] == 'Applied Sc':
    df1_collegenamelist.append('Cyber College')
  elif df1_majorlist[i] == 'Accounting' or df1_majorlist[i] == 'Business Admin' or df1_majorlist[i] == 'Economics':
    df1_collegenamelist.append('College of Business')
  elif df1_majorlist[i] == 'Elementary Ed' or df1_majorlist[i] == 'Secondary Ed':
    df1_collegenamelist.append('College of Education')
  elif df1_majorlist[i] == 'Biology' or df1_majorlist[i] == 'Chemistry' or df1_majorlist[i] == 'English':
    df1_collegenamelist.append('College of Art and Science')
  else:
    df1_collegenamelist.append('NULL')

#df1_collegenamelist
#len(df1_collegenamelist)

df1.insert(3,'Name of Colleges',df1_collegenamelist)
df1
```

Encoding: Format Integration (Semester)

Imputing the Null values

```python
# Month Day
cdf = concat_df.copy(deep=True)
for i in range(len(cdf)):
  if pd.isna(cdf.loc[i,'Month Day']):
    if cdf.loc[i,'Semester'] == "Fall":
      cdf.loc[i,'Month Day'] = "Dec. 15"
    elif cdf.loc[i,'Semester'] == "Spring":
      cdf.loc[i,'Month Day'] = "May. 15"
    elif cdf.loc[i,'Semester'] == "Summer 1":
      cdf.loc[i,'Month Day'] = "June 1"
    elif cdf.loc[i,'Semester'] == "Summer 2":
      cdf.loc[i,'Month Day'] = "July 4"
    elif cdf.loc[i,'Semester'] == "Summer 3":
      cdf.loc[i,'Month Day'] = "July 15"
    elif cdf.loc[i,'Semester'] == "Summer 4":
      cdf.loc[i,'Month Day'] = "August 15"
    else:
      cdf.loc[i,'Month Day'] = "NULL"

#Name of Old_Colleges
for i in range(len(cdf)):
  if pd.isna(cdf.loc[i,'Name of Old_Colleges']):
    if cdf.loc[i,'Name of Colleges'] == "Cyber College":
      cdf.loc[i,'Name of Old_Colleges'] = "Computer Communications"
    elif cdf.loc[i,'Name of Colleges'] == "College of Business":
      cdf.loc[i,'Name of Old_Colleges'] = "College of Business and IS"
    elif cdf.loc[i,'Name of Colleges'] == "College of Education":
      cdf.loc[i,'Name of Old_Colleges'] = "College of Education"
    elif cdf.loc[i,'Name of Colleges'] == "College of Art and Science":
      cdf.loc[i,'Name of Old_Colleges'] = "College of Science"
    else:
      cdf.loc[i,'Name of Old_Colleges'] = "NULL"
temp_df= cdf.copy(deep = True)
cdf
```
